## 결측값 처리
```
df.dropna(axis=1) # axis=0: 행지워라, axis=1: 열지워라
df.fillna()
df.isna()
pd.isna(df)
```

````python
# 결측값 지워라 내부적으로
dfcp = df.copy()
dfcp.dropna(axis=0, inplace=True)
dfcp
````

````python
# 결측값 지워서 덮어씌우기 행기준
dfcp = df.copy()
dfcp = dfcp.dropna(axis=0)
dfcp
````

````python
# 결측값 지워서 덮어씌우기 열기준
dfcp = df.copy()
dfcp = dfcp.dropna(axis=1)
dfcp
````

````python
# 결측을 9999999로
dfcp = df.copy()
dfcp = dfcp.fillna(9999999)
dfcp
````

````python
# MGR의 결측은 777777로 COMM의 결측은 888888로
dfcp = df.copy()
dfcp['COMM'].fillna(888888, inplace=True)
dfcp['MGR'].fillna(777777, inplace=True)
dfcp
````

````python
# 특정 컬럼의 결측값을 999로 바꾸고 타입도 숫자로
dfcp = df.copy()
dfcp[['MGR', 'COMM']] = dfcp[['MGR', 'COMM']].fillna(999).astype('int')
dfcp.info()
````

````python
# 결측이냐?
dfcp = df.copy()
# pd.isna(dfcp)
dfcp.isna()
# dfcp.isnull()
````

````python
# 값있냐?
dfcp = df.copy()
dfcp.notna()
#dfcp.notnull()
````
