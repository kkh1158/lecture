## subplots(nrows, ncols) vs. subplot(행, 열, 인덱스)

<img src="https://wikidocs.net/images/page/141537/subplot_03.png" width=700>

* subplot(행, 열, 인덱스,     figsize=(4,2))<br><br>
* df.plot(kind="bar", figsize=(2,2))   <br><br>
* plt.figure(figsize=(4,2)) <br>
* subplots() <br><br>

````python
x = ['apple', 'orange', 'lemon', 'lime']
y = [10, 15, 5, 20]

plt.figure(figsize=(12,4))

plt.subplot(1,3,1)
plt.bar(x, y)

plt.subplot(1,3,2)
plt.scatter(x, y)

plt.subplot(1,3,3)
plt.plot(x, y)

plt.show()
````

````python
x = ['apple', 'orange', 'lemon', 'lime']
y = [10, 15, 5, 20]

fig, ax = plt.subplots(1, 3, figsize=(12,4))  #------------1줄 3개칸 
ax[0].bar(x, y)
ax[1].scatter(x, y)
ax[2].plot(x, y)
plt.show()
````

# 옵션

## 러그(rug)
* 데이터 위치를 x축 위에 작은 선분(rug)으로 나타내어 실제 데이터들의 위치 표시

````python
tips = sns.load_dataset("tips")
print('x:', tips['total_bill'].dtype)
print('col:', tips['time'].dtype)
print('hue:', tips['smoker'].dtype)
sns.displot(data=tips, kind="ecdf", x="total_bill", col="time", hue="smoker", rug=True)
plt.show()
````

# 1차원 분포 플롯
* 연속형 ㅡ rugplot, kdeplot, distplot
* 범주형 ㅡ countplot

## displot: 대장인데 구짐

https://seaborn.pydata.org/generated/seaborn.displot.html#seaborn.displot


* histplot
* kedplot
* ecdfplot

````python
print('x:', tips['total_bill'].dtype)
sns.displot(data=tips, x="total_bill", kde=True)
plt.show()
````

## histplot: 히스토그램

https://seaborn.pydata.org/generated/seaborn.histplot.html#seaborn.histplot

* kde=True # 커널밀도도 같이

````python
print('x:', tips['total_bill'].dtype)
plt.figure(figsize=(12,4))
plt.subplot(1,3,1)
sns.histplot(data=tips, x="total_bill")
plt.subplot(1,3,2)
sns.histplot(data=tips, x="total_bill", kde=True)
plt.subplot(1,3,3)
sns.histplot(data=tips, x="total_bill", kde=True, log_scale=True)
plt.show()
````

## kdeplot: 커널 밀도 추정(kernel density Estimation)

https://seaborn.pydata.org/generated/seaborn.kdeplot.html#seaborn.kdeplot

* log_scale=True # 정규분포화

````python
print('x:', tips['total_bill'].dtype)
plt.figure(figsize=(12,4))
plt.subplot(1,3,1)
sns.kdeplot(data=tips, x="total_bill", log_scale=False)
plt.subplot(1,3,2)
sns.kdeplot(data=tips, x="total_bill", log_scale=True)
plt.subplot(1,3,3)
sns.kdeplot(np.log(tips['total_bill']))
plt.show()
````

## countplot

https://seaborn.pydata.org/generated/seaborn.countplot.html?highlight=count


````python
titanic = sns.load_dataset("titanic")
print('x:', titanic['class'].dtype)
print('hue:', titanic['sex'].dtype)
plt.figure(figsize=(12,8))
plt.subplot(2,2,1)
sns.countplot(x=titanic["class"])
plt.subplot(2,2,2)
titanic['class'].value_counts().sort_index().plot(kind='bar')
plt.subplot(2,2,3)
sns.countplot(x=titanic["class"], hue=titanic['sex'])
plt.subplot(2,2,4)
titanic[['class','sex']].value_counts().sort_index().plot(kind='bar')
plt.show()
````

````python
titanic = sns.load_dataset("titanic")
print('x:', titanic['class'].dtype)
plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
sns.countplot(x=titanic["class"])
plt.subplot(1,2,2)
titanic['class'].value_counts().sort_index().plot(kind='bar')
plt.show()
````

## barplot: x별 y의 평균
https://seaborn.pydata.org/generated/seaborn.barplot.html

````python
print('x:', titanic['class'].dtype)
print('y:', titanic['survived'].dtype)
plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
sns.barplot(data=titanic, x="class", y='survived', estimator='mean', errorbar=('ci', 95))
plt.subplot(1,2,2)
titanic.groupby('class')['survived'].mean().sort_index().plot(kind='bar')
plt.show()
````

# 2차원 실수형 데이터

## scatterplot

https://seaborn.pydata.org/generated/seaborn.scatterplot.html

````python
penguins = sns.load_dataset("penguins")
print('x:', penguins['bill_length_mm'].dtype)
print('y:', penguins['bill_depth_mm'].dtype)
print('hue:', penguins['species'].dtype)
sns.scatterplot(data=penguins, x="bill_length_mm", y="bill_depth_mm", hue="species")
plt.show()
````

## jointplot

https://seaborn.pydata.org/generated/seaborn.jointplot.html?highlight=ointplot#seaborn.jointplot

* 스캐터 플롯 + 가장자리(margin)에 히스토그램 표시
* kind='scatter'  kind='kde'

````python
print('x:', penguins['bill_length_mm'].dtype)
print('y:', penguins['bill_depth_mm'].dtype)
print('hue:', penguins['species'].dtype)
sns.jointplot(data=penguins, x="bill_length_mm", y="bill_depth_mm", hue="species", kind='scatter')
plt.show()
````
