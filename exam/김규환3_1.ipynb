```python
import numpy as np
import pandas as pd
```

```python
car = pd.read_csv('./mtcars.csv')
car.info()
```

```python
car
```

```python
# qsec 컬럼을 최소최대 척도(Min-Max Scale)로 변환 qsec_mm
from sklearn.preprocessing import MinMaxScaler

minmax = MinMaxScaler()
minmax.fit(car['qsec'].values.reshape(-1,1))
res = minmax.transform(car['qsec'].values.reshape(-1,1))
car['qsec_mm'] = res
car
```

```python
# 1.2. qsec_mm 피쳐값이 0.5보다 큰 값을 갖는 레코드 수
car[car['qsec_mm']>0.5].shape[0]
```

```python
# carb 피쳐값에 대해서 25%, 50%, 75% 값 출력
car['carb'].describe()['25%'], car['carb'].describe()['50%'], car['carb'].describe()['75%']
```

```python
np.quantile(car['carb'], q=0.25)
```

```python
# disp로 그룹핑한 그룹별 mpg 평균값의 90퍼값
car.groupby('disp')['mpg'].apply(np.mean).quantile(0.9)
```

```python
car['disp'].sort_values().index[:10]
```

```python
## disp 피쳐값에 대해서 오름차순 정렬 시 10번째 값으로 1~10위의 disp 값 변경
disp10번째값 = car['disp'].sort_values().values[9] # 140.8
상위10개인덱스 = car['disp'].sort_values().index[:10] # [19, 18, 17, 25, 27, 2, 20, 26, 31, 8]
car.loc[상위10개인덱스,'disp'] = disp10번째값
car['disp'].sort_values()
```

```python
## disp 피쳐값에 대해서 80이상의 값에 대한 평균 산출
car[car['disp']>=80]['disp'].mean()
```

```python
disp80퍼값
```

```python
disp80퍼값 =  car['disp'].quantile(0.8)
car[car['disp']>=disp80퍼값]['disp'].mean()
```

```python
## mpg 피쳐값에 대해서 80% 량의 데이터를 추출 후 'mpg'가 20 이상인 값을 중앙값(median)으로 바꾸고 변경전, 변경 후 'mpg'의 표준편차 출력
from sklearn.model_selection import train_test_split

중앙값 = car['mpg'].median() # 19.2
추출, 나머지 = train_test_split(car['mpg'], train_size=0.8)
수정할인덱스 = 추출[추출.values>=20].index
변경전표준편차 = 추출[추출.values>=20].std()
추출[수정할인덱스] = 중앙값
변경후표준편차 = 추출.std()
변경전표준편차, 변경후표준편차
```
